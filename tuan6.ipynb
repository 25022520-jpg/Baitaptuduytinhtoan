Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của key đó 

```python
line=input()
pairs=line.split()
res={}
for item in pairs:
    key,value=item.split(":")
if key not in res:
    res[key]=[value]
else:
    res[key].append(value)
print(res)
```

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của
key. 


```python
def nhap_dict():
    chuoi=input()
    d={}
    cac_cap=chuoi.split()
    for cap in cac_cap:
        if ":" in cap:
            key,value=cap.split(":", 1)
            d[key.strip()]=int(value.strip())
    return d
dict1=nhap_dict()
print(f"Nhap dictionary 1:{dict1}") 
dict2=nhap_dict()
print(f"Nhap dictionary 2:{dict2}") 
result=dict1.copy()
for key,value in dict2.items():
    if key in result:
        result[key]+=value
    else:
        result[key]=value

print("Ket qua sau khi cong 2 dictionary:",result)
```

W6A10

Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
trong tuple.

```python
lst=list(map(int,input().split()))
k=int(input())
n=len(lst)
res=[]
for i in range(n):
    for j in range(i+1,n):
        if lst[i]+lst[j]==k:
            res.append((lst[i], lst[j]))
unique_res=list(set(res))
unique_res.sort()
print(unique_res)
```

[Tách số chẵn và lẻ từ tuple]
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.

```python
input_data=input("Nhập các số (cách nhau bởi dấu cách): ")
number=[int(x) for x in input_data.split()]
even_list=[]
odd_list=[]
for n in number:
    if n % 2==0:
        even_list.append(n)
    else:
        odd_list.append(n)
print("Các số chẵn:", tuple(even_list))
print("Các số lẻ:", tuple(odd_list))
```

[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
trong số đó. 

```python
input_data=input()
number=[int(x) for x in input_data.split()]
counts={}
for num in number:
    counts[num]=counts.get(num,0)+1
max_counts=-1
res=None
for num in sorted(counts.keys()):
    if counts[num]>max_counts:
        max_counts=counts[num]
        res=num
print(res)
```

dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In

[Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp keyvalue từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới. 


```python
n=input().split()
k=int(input())
d={}
l={}
a=len(n)
for i in range(0,a,2):
    d[n[i]]=n[i+1]
for key in sorted(d.keys()):
    if int(d[key])>k:
        l[key]=d[key]
print(l)
```
