Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
import math
a = float(input("Nhap canh a: "))
b = float(input("Nhap canh b: "))
c = float(input("Nhap canh c: "))
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        loai = "Tam giac Deu"
    elif a == b or b == c or a == c:
        if (a**2 + b**2 == round(c**2, 2)) or (a**2 + c**2 == round(b**2, 2)) or (b**2 + c**2 == round(a**2, 2)):
            loai = "Tam giac Vuong Can"
        else:
            loai = "Tam giac Can"
    elif (a**2 + b**2 == round(c**2, 2)) or (a**2 + c**2 == round(b**2, 2)) or (b**2 + c**2 == round(a**2, 2)):
        loai = "Tam giac Vuong"
    else:
        loai = "Tam giac Thuong"
    chu_vi = a + b + c
    p = chu_vi / 2  
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print("Loai tam giac:", loai)
    print("Chu vi:", chu_vi)
    print("Dien tich:", round(dien_tich, 2))

else:
    print("Khong la ba canh cua mot tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a = float(input())
b = float(input())
c = float(input())
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm.")
        else:
            print("Phương trình vô nghiệm.")
    else:
        x = -c / b
        print("Phương trình có nghiệm duy nhất x =", x)
else:
    delta = b**2 - 4*a*c

    if delta < 0:
        print("Phương trình vô nghiệm thực.")
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép x =", x)
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có 2 nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
```

Bài 8: Giải hệ phương trình tuyến tính:
ax  by  m
cx  dy  n

```python
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
m = float(input("Nhập m: "))
n = float(input("Nhập n: "))
D = a*d - b*c
if D != 0:
    x = (m*d - b*n) / D
    y = (a*n - m*c) / D
    print("Hệ có nghiệm duy nhất:")
    print("x =", x)
    print("y =", y)
else:
    if a*n == m*c and b*n == m*d:
        print("Hệ có vô số nghiệm.")
    else:
        print("Hệ vô nghiệm.")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x0 = float(input())
y0 = float(input())
R = float(input())
x = float(input())
y = float(input())
d2 = (x - x0)**2 + (y - y0)**2
if d2 == R**2:
    print("Điểm A nằm trên đường tròn.")
elif d2 < R**2:
    print("Điểm A nằm trong đường tròn.")
else:
    print("Điểm A nằm ngoài đường tròn.")
```
